```python
#QUESTION 2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = '/content/Dataset .csv'
df = pd.read_csv('/content/Dataset .csv')


rows, columns = df.shape
print(f'The dataset has {rows} rows and {columns} columns.')


missing_values = df.isnull().sum()
print('Missing values in each column:\n', missing_values)


numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())


df = df.dropna()


basic_stats = df.describe().T
print('Basic statistical measures:\n', basic_stats)


categorical_cols = ['Country Code', 'City', 'Cuisines']

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.show()


top_cuisines = df['Cuisines'].value_counts().head(10)
print('Top 10 Cuisines with the highest number of restaurants:\n', top_cuisines)


top_cities = df['City'].value_counts().head(10)
print('Top 10 Cities with the highest number of restaurants:\n', top_cities)
```

```python
#QUESTION 3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap


file_path = '/content/Dataset .csv'
df = pd.read_csv('/content/Dataset .csv')


rows, columns = df.shape
print(f'The dataset has {rows} rows and {columns} columns.')

missing_values = df.isnull().sum()
print('Missing values in each column:\n', missing_values)


numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())


df = df.dropna()


basic_stats = df.describe().T
print('Basic statistical measures:\n', basic_stats)



for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.show()


top_cuisines = df['Cuisines'].value_counts().head(10)
print('Top 10 Cuisines with the highest number of restaurants:\n', top_cuisines)


top_cities = df['City'].value_counts().head(10)
print('Top 10 Cities with the highest number of restaurants:\n', top_cities)


map_restaurants = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=12)


for _, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name'],
    ).add_to(map_restaurants)


map_restaurants.save('restaurants_map.html')


map_restaurants


plt.figure(figsize=(10, 6))
sns.countplot(y='City', data=df, order=df['City'].value_counts().index)
plt.title('Distribution of Restaurants across Cities')
plt.show()


plt.figure(figsize=(10, 6))
sns.countplot(y='Country Code', data=df, order=df['Country Code'].value_counts().index)
plt.title('Distribution of Restaurants across Countries')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', hue='Aggregate rating', data=df, palette='coolwarm')
plt.title('Correlation between Location and Rating')
plt.show()


correlation_matrix = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
print('Correlation matrix:\n', correlation_matrix)
```
